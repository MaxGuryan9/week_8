## 1. What makes a model "linear"? "Linear" in what?

A model is linear because of the exclusive interaction between a variable $X_n$ and its $\beta_n$ parameter. The parameter coefficient $\beta_n$ is always multiplied with its respective $X_n$, never raised to an exponent or multiplied by another parameter. Essentially, everytime there is a change in $X_n$, the output will be changed by $\Delta X_n*\beta_n$. If you have the equation $\hat{y} = \beta_0 + \beta_1 x + \beta_2 x_2 + ... + \beta_n x_n$, each term will have its own linear impact on the model depending on its $\Delta X_n*\beta_n$ value. Ultimately, creating a linear predictive model. It is also important to note that the linearity comes from the parameters, not the features. You could add a term $X^2_4$ and as long as it is multiplied by its parameter and correctly implemented into $\hat{y} = \beta_0 + \beta_1 x + \beta_2 x_2 + ... + \beta_n x_n$ then the model itself is still linear. 

## 2. How do you interpret the coefficient for a dummy/one-hot-encoded variable?

First, the intercept of a model is the output of the model when all other terms are 0. In the context of dummy/one-hot-encoded variables, this is important in order to avoid an unexplainable $\beta_0$ value. Lets say we have the column `color`, with the possible values of color being `["Red", "Green", "Blue"]`, and we one-hot-encode it. This creates 3 binary columns where 1 denotes the color present and 0 denotes the color(s) not present. In this scenario the intercept is unexplainable because there are three columns and three possible colors. If all three terms were 0, then color would be N/A, and we can't have $\beta_0 = N/A$. However, by removing one of the possible categories, "Red" for the sake of this example, we give $\beta_0$ meaning. There are only three options, and all instances must fit into one of three categories. If "Green" = 0 and "Blue" = 0, you know that instance is "Red". Thus,  $\beta_0$, the intercept, is "Red", ultimately giving $\beta_0$ meaning. Overall, in this example, the coefficients tell you how likely a specified color output is in comparison to "Red" while all other features are held constant. If $\beta_{\text{green}} = 5$ then a green instance is 5 units higher than a red instance.

## 3. Can linear regression be used for classification? Explain why, or why not.

In theory, a linear regression model could be used for classification, but it is unrealistic and requires additional steps. Linear regresion cannot put each instance into a specified category as an output of the function because outputs are unbounded and continuous. The output of the model can potentially be used to make a categorization, but that would require extra steps to calibrate new parameters where a given range denotes a given category. It would be much more beneficial to use a logistic regression, where the outputs can be understood at probabilities, making it slightly easier to classify outputs. Finally, a typical relationship between a feature and a binary outcome is usually S-shaped. Near the extremes the probability changes slowly, but changes more rapidly near the middle. This abnormal shape is not well captured by a linear regression line.

## 4. If you have a high accuracy on training but low on testing, what might be the problem? It also might be you see a pattern in the residuals.

First, lets look at high accuracy on training but low on testing. This means that the model has overfit itself to the training data, and cannot use those patterns to accurately predict the test data. Metaphorically, it has memorized the correct output for each input in the training dataset, but struggles to apply this knowledge to similar conceptual questions in the testing dataset. Second, if you see a pattern in the residuals this means that the model is consistenly over/underfitting to the data in different areas of the dataset. This typically means that the regression is not linear and a non-linear regression model is going to better fit the trends of the data.

1. You can add features with increasing degrees. If you have data that seems to have a cubic trend, you can use `PolynomialFeatures` to create new columns for each existing feature with increasing degree. If we see `degree = 3` each feature will have a new column with $X^2$ and $X^3$. Each of these new features will have its own $\beta$-value allowing the overall model to be linear, but still better fit the cubic-like shape of the data. It is important to not have too many degrees as this will lead to extreme overfitting of the models. 
2. You can create bins that span equal sized ranges of the dataset. Once categorized, this will become a new feature of the dataset that will be one-hot-encoded, and given its respective data $\beta$-value. This allows each bin to have influence on the predicted output of the model, helping the overall model fit the data better.

## 6. What is the interpretation of the intercept? A slope coefficient for a variable? The coefficient for a dummy/one-hot-encoded variable?

The intercept is the predictive output of the model when all features are equal to 0. Essentially, if there was no feature influence on the output, what is the predicted output of the model. A slope coefficient for a variable is the weighted influence that feature will have on the predicted output of the model, assuming all else is equal. A feature with a slope coefficient of 0.5 will have less of an influence on the predicted output compared to a feature with a slope coefficient of 5.
